```python
# GESTIONAMOS LOS NULOS

columnas_unknown = ["índice de igualdad legal", "legalidad_colectivo_lgbtiq+", 'fecha_despenalización', 'constitución', 'protección_amplia', 'empleo', 'crimen_odio', 'incitación', 'prohibición_terapias', 'matrimonio_igualitario', 'uniones_civiles', 'adopción_conjunta', 'adopción_segundo_progenitor']   

# iteramos por la lista de columnas a las que le vamos a cambiar los nulos por "Uknown"
for columna in columnas_unknown:
    
    # reemplazamos los nulos por el valor Unknown para cada una de las columnas de la lista
    df_unido2[columna] = df_unido2[columna].fillna("Unknown")

print("Después del reemplazo usando 'fillna' quedan los siguientes nulos")
df_unido2[columnas_unknown].isnull().sum()
```

```python
df_unido2[df_unido2['fecha_despenalización'] == 'Unknown']['país']
```

```python
# PARA PODER PASAR LA COLUMNA A INT, LE PEDÍ A CHAT QUE ME DIJERA LA INFO DE LOS PAISES FALTANTES, PORQ NO ME DEJABA CON 'Unknown'
despenalizacion = {
    'Bosnia and Herzegovina': 1998,
    'Brunei': 2007,  
    'Central African Republic': 1996,
    'Cote d\'Ivoire': 2016,
    'Czechia': 1962,
    'Democratic Republic of Congo': 0,  # Este país sigue teniendo la homosexualidad ilegal
    'Indonesia': 0,  # La homosexualidad no está penalizada a nivel nacional, pero depende de la región
    'Iraq': 0,  # Homosexualidad sigue siendo ilegal en Iraq
    'Saint Kitts and Nevis': 1998,
    'Saint Vincent and the Grenadines': 2001,
    'Sao Tome and Principe': 2012,
    'Taiwan': 1997
}

# Reemplazar los valores 'Unknown' en 'fecha_despenalización' con los años del diccionario
for pais, ano in despenalizacion.items():
    df_unido2.loc[(df_unido2['país'] == pais) & (df_unido2['fecha_despenalización'] == 'Unknown'), 'fecha_despenalización'] = ano
```

```python
# Vemos los países que tienen Unknown en la columna "índice de igualdad legal"
df_unido2[df_unido2['índice de igualdad legal'] == 'Unknown']['país']
```

```python
# PARA PODER PASAR LA COLUMNA A FLOAT, LE PEDÍ A CHAT QUE ME DIJERA LA INFO DE LOS PAISES FALTANTES, PORQ NO ME DEJABA CON 'Unknown'
indice = {
    'Bosnia & Herzegovina': 30,
    'Brunei Darussalam': 0,
    'Cabo Verde': 50,
    'Czech Republic': 80,
    'Côte d\'Ivoire': 50,
    'Kosovo': 40,
    'Micronesia': 60,
    'Palestine': 10,
    'Saint Kitts & Nevis': 10,
    'Saint Vinc. & the Gren.': 20,
    'Sao Tome & Principe': 60,
    'Vatican City': 0}


# Reemplazar los valores 'Unknown' en 'índice de igualdad legal' con los años del diccionario
for pais, ind in indice.items():
    df_unido2.loc[(df_unido2['país'] == pais) & (df_unido2['índice de igualdad legal'] == 'Unknown'), 'índice de igualdad legal'] = ind
```

```python
# Cambiamos el tipo de datos de la columna "índice de igualdad legal"
df_unido2['índice de igualdad legal'] = df_unido2['índice de igualdad legal'].astype(float)
```

```python
# Cambiamos el tipo de datos de la columna "año_actualización_datos"
df_unido2['fecha_despenalización'] = df_unido2['fecha_despenalización'].astype(int)
```
