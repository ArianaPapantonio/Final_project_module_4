```python
# HANDLE MISSING VALUES

# List of columns where missing values will be replaced with "Unknown"
columns_unknown = [
    "índice de igualdad legal", "legalidad_colectivo_lgbtiq+", "fecha_despenalización",
    "constitución", "protección_amplia", "empleo", "crimen_odio", "incitación",
    "prohibición_terapias", "matrimonio_igualitario", "uniones_civiles",
    "adopción_conjunta", "adopción_segundo_progenitor"
]

# Iterate over the list of columns and replace missing values with "Unknown"
for column in columns_unknown:
    df_merged_2[column] = df_merged_2[column].fillna("Unknown")

# Display how many missing values remain in those columns after replacement
print("After using 'fillna', the following nulls remain:")
df_merged_2[columns_unknown].isnull().sum()
```

```python
# To convert the column to integers, we needed to remove "Unknown" values
# This dictionary assigns decriminalization years to specific countries with missing data
decriminalization_years = {
    'Bosnia and Herzegovina': 1998,
    'Brunei': 2007,  
    'Central African Republic': 1996,
    "Cote d'Ivoire": 2016,
    'Czechia': 1962,
    'Democratic Republic of Congo': 0,  # Still criminalizes homosexuality
    'Indonesia': 0,  # Depends on region; not penalized nationwide
    'Iraq': 0,  # Homosexuality is still illegal
    'Saint Kitts and Nevis': 1998,
    'Saint Vincent and the Grenadines': 2001,
    'Sao Tome and Principe': 2012,
    'Taiwan': 1997
}

# Replace 'Unknown' values in 'fecha_despenalización' with the correct year from the dictionary
for country, year in decriminalization_years.items():
    df_merged_2.loc[
        (df_merged_2['país'] == country) & (df_merged_2['fecha_despenalización'] == 'Unknown'),
        'fecha_despenalización'
    ] = year
```

```python
# Display the countries that have "Unknown" in the "índice de igualdad legal" column
df_merged_2[df_merged_2['índice de igualdad legal'] == 'Unknown']['país']
```

```python
# To convert the column to float, we needed to remove "Unknown" values
# This dictionary provides the legal equality index values for countries with missing data
equality_index = {
    'Bosnia & Herzegovina': 30,
    'Brunei Darussalam': 0,
    'Cabo Verde': 50,
    'Czech Republic': 80,
    "Côte d'Ivoire": 50,
    'Kosovo': 40,
    'Micronesia': 60,
    'Palestine': 10,
    'Saint Kitts & Nevis': 10,
    'Saint Vinc. & the Gren.': 20,
    'Sao Tome & Principe': 60,
    'Vatican City': 0
}

# Replace 'Unknown' values in 'índice de igualdad legal' with the values from the dictionary
for country, index in equality_index.items():
    df_merged_2.loc[
        (df_merged_2['país'] == country) & (df_merged_2['índice de igualdad legal'] == 'Unknown'),
        'índice de igualdad legal'
    ] = index
```

```python
# Convert the data type of the "índice de igualdad legal" column to float
df_merged_2['índice de igualdad legal'] = df_merged_2['índice de igualdad legal'].astype(float)
```

```python
# Convert the data type of the "fecha_despenalización" column to integer
df_merged_2['fecha_despenalización'] = df_merged_2['fecha_despenalización'].astype(int)
```
